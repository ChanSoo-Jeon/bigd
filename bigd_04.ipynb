```r
#6 제시된 Pre_Season_Batter_utf.csv 파일을 fread()를 사용하여
# DF 이름의 tibble로 읽어 들인 후 확인하세요.
# (Hint) fread(), %>%, as_tibble(), show()

DF <- fread("/content/Pre_Season_Batter_utf.csv",header = T, encoding = "UTF-8") %>% as_tibble()
DF %>% show()
```

```r
#7 sample() 함수를 이용하여 1부터 100까지 숫자들 중 10개를 추출하여
# plot() 그래프를 그리세요.

sample(1:100,10,replace=FALSE) %>% plot()
```

```r
#8 DF 데이터에서 G(출전게임수)가 증가함에 따라 avg(타율)이
# 어떠한 관계를 갖는지 plot() 그래프로 확인한 후 설명하세요.

plot(DF$G,DF$avg,main="출전게임수에 따른 타율")
# 경기수와 타율은 상관관계가 없다.
```

```r
#9 DF 데이터에서 AB(타수(타석-볼넷-사구-희생번트-희생플라이)),
# H(안타(1루타+2루타+3루타 +홈런))이 어떠한 관계를 갖는지 plot()
# 그래프로 확인한 후 설명하세요.

plot(DF$AB,DF$H)
#타석이 많을수록 안타 개수가 늘어나는 비례관계를 갖는다.
```

```r
#10 DF 데이터의 5열부터 9열까지의 plot() 그래프를
# 한 번에 그린 후 양의 AB(타수)와 R(득점) 의 관계를 설명하세요.

DF %>% select(5:9) %>% plot()
# 타수가 늘어나면 득점도 늘어나는 비례 관계를 갖는다.
```

```r
#11 DF 데이터에서 DF 데이터에서 숫자형 데이터들을 하나의 plot() 그래프에 나타내세요.
DF %>% select_if(is.numeric) %>% plot()
```

```r
#13 DF 데이터에서 "KIA", "SK", "LG" 팀의 year(년도)별
# avg_mean(평균 타율)을 team_year_avg에 저장한 후 확인하세요.
# (Hint) year(년도), avg(타율), group_by(), summarise(), filter()

team_year_avg <- DF %>% filter(team %in% c("KIA","SK","LG")) %>% group_by(team,year) %>%
  summarise(avg_mean = mean(avg,na.rm=TRUE))

team_year_avg %>% show()
```

```r
#14 team_year_avg에서 "KIA"팀의 정보는 kia,
# "SK"팀의 정보는 sk, "LG"팀의 정보는 lg에 저장 하세요.

kia <- team_year_avg %>% filter(team == "KIA")
sk <- team_year_avg %>% filter(team == "SK")
lg <- team_year_avg %>% filter(team == "LG")

kia %>% show()
sk %>% show()
lg %>% show()
```

```r
#15 아래의 스크립트가 의미하는 것이 무엇인지 설명하고 그래프를 그려 확인하세요.

plot(x = team_year_avg$year, y = team_year_avg$avg_mean, type = "n", main = "Team batting average by Year", xlab = "YEAR", ylab = "AVG")
points(x = lg$year, y = lg$avg_mean, col = "purple", pch = 8, cex = 1)
lines(x = lg$year, y = lg$avg_mean, col = "purple")
points(x = sk$year, y = sk$avg_mean, col = "green", pch = 11, cex = 1)
lines(x = sk$year, y = sk$avg_mean, col = "green")
points(x = kia$year, y = kia$avg_mean, col = "orange", pch = 4, cex = 1)
lines(x = kia$year, y = kia$avg_mean, col = "orange")
legend("bottomright", legend = c("LG","SK","KIA"), pch = c(8,11,4), col = c("purple", "green", "orange"), cex = 1.5)
```

```r
#16 DF 데이터에서 AB(타수), R(득점)이 어떠한 관계를 갖는지
# geom_point() 그래프로 확인한 후 설명하세요. 로그 스케일을 적용해서도 확인하세요.

DF %>% ggplot(aes(x=AB,y=R)) + geom_point() + scale_x_log10()
```

```r
#17 DF 데이터에서 AB(타수), R(득점)을
# 아래의 조건에 맞춰 geom_point() 그래프로 작성하세요.
#A. team(팀) 별로 서로 다른 색으로 표시
#B. R(득점)과 동그란 포인트의 사이즈가 비례하도록 표시
#C. 투명도는 0.5를 적용

DF %>% ggplot(aes(x=AB,y=R),size=R,col=team) + geom_point(alpha=0.5)
```

```r
#19 DF 데이터에서 team(팀)별 R(득점)의 합, H(안타)의 합을
# team_R_H_sum에 저장하고 확인 하세요.

team_R_H_sum <- DF %>% group_by(team) %>% summarise(R_sum=sum(R),H_sum=sum(H))
team_R_H_sum %>% show()
```

```r
#20
team_R_H_sum %>% ggplot(aes(x=team,y=R_sum)) +
  geom_bar(stat = "identity",aes(reorder(team,-R_sum),fill=team))
```

```r
#21
team_R_H_sum %>% ggplot(aes(x=team,y=R_sum)) +
  geom_bar(stat = "identity",aes(reorder(team,-R_sum),fill=team)) +
  coord_flip()
```
