```r
#11 dplyr패키지를 이용하여 DF에서 team(소속팀)이 ‘한화’이고 G(출전게임수)가 5회 이상, AB
#(타수(타석-볼넷-사구-희생번트-희생플라이))가 15회 이상인 선수들의 team, batter_name,
#G, AB를 출력하세요. 단, 중복 값은 제거하고 고유한 값만을 출력하세요. (Hint) unique()

show(unique(DF %>% filter(team == "한화" & G >= 5 & AB >= 15) %>% select(team,batter_name,G,AB)))
```

```r
#12 dplyr패키지를 이용하여 DF에서 team(소속팀)을 group으로 묶어 team(소속팀)별 평균 G(출
# 전게임수), 평균 AB(타수(타석-볼넷-사구-희생번트-희생플라이)), 합계 R(득점)을 출력

DF %>% group_by(team) %>% summarise(mean_G = mean(G), mean_AB = mean(AB), total_R = sum(R))
```

```r
#13 dplyr패키지를 이용하여 DF에서 team(소속팀)과 batter_name(타자 이름)을 group으로 묶어
# 소속팀의 타자별 평균 G(출전게임수), 평균 AB(타수(타석-볼넷-사구-희생번트-희생플라
# 이)), 합계 R(득점)을 출력하세요
DF %>% group_by(team,batter_name) %>% summarise(mean_G = mean(G),mean_AB=mean(AB),total_R=sum(R))
#na.rm=T
```

```r
#14 dplyr패키지를 이용하여 DF를 team(소속팀)은 오름차순으로, avg(타율)은 내림차순을 기준
# 으로 정렬하여 team(소속팀), batter_name(타자 이름), avg(타율)만 출력하세요.

DF %>% arrange(team,desc(avg))  %>% select(team,batter_name,avg)
```

```r
#15 DF에 존재하는 team(소속팀)을 중복 없이 오름차순으로 정렬하여 출력하세요.

DF %>% select(team) %>% unique() %>% arrange(team) %>% show()
```

```r
#16 DF에서 team(소속팀)이 ‘히어로즈’인 선수들의 batter_name(타자 이름)과
# avg(타율)을 avg(타율)이 높은 순으로 중복 없이 정렬하여 출력하세요.

DF %>% filter(team == "히어로즈") %>% select(batter_name,avg) %>% unique() %>% arrange(desc(avg)) %>%show()
```

```r
#17 DF에서 batter_id(타자의 고유 아이디)는 ‘ID’, batter_name(타자 이
# 름)은 ‘이름’, year(년도)는 ‘년도’라는 이름으로 변경하여 DF1에 저장한 후
# DF1과 DF를 출력 해 비교하세요.
DF1 <- DF %>% select(ID = batter_id,이름=batter_name,년도=year)
show(DF1)
show(DF)
```

```r
#18 DF에서 batter_id(타자의 고유 아이디)를 batter_id(타자의 고유 아이디)
# +100으로 변경하여 DF1에 저장한 후 DF1과 DF를 출력하여 비교하세요.

DF1 <- DF %>% mutate(batter_id = batter_id + 100)
show(DF1)
show(DF)
```

```r
#20 DF에 경기 당(G) 득점(R)에 해당하는 G_R 열을 생성한 후 batter_name(타자 이름)과 G(출전게임수),
# R(득점), G_R(경기당 득점)에 해당하는 열만 출 력하세요.
DF %>% mutate(G_R=R/G) %>% select(batter_name,G,R,G_R) %>% show()
```

```r
#21 DF에서 비복원 추출 방식으로 30개의 샘플을 추출해 train_set에
# 장하고 출력하세요. (Hint) sample_n(), replace=F 비복원추출

train_set <- sample_n(DF,30,replace=F)
train_set %>% show()
```

```r
#22 DF에서 비복원 추출 방식으로 70%의 샘플을 추출해 train_set에
# 장하고 출력하세요. (Hint) sample_frac()

train_set <- sample_frac(DF,0.7)
train_set %>% show()
```

```r
# 23  DF에서 위 문제에서 구한 train_set을 제외한 샘플을 추출하여
# test_set에 저장하고 출력하세요. (Hint) setdiff()

test_set <- setdiff(DF,train_set)
test_set %>% show()
```
